## Types de polarisation de la lumière ##

### Polarisation verticale ###

Si E_0x = 0 alors l'onde lumineuse n'a qu'une seule composante en y et est verticale.

![image](polarisation_0.png)

### Polarisation à 45° ###

S'il n'y a pas de différence de phase et que E_0x = E_0y alors E_x = E_y.

![image](polarisation_45.png)

### Polarisation circulaire ###

Si la différence de phase est de 90° et si E_0x = E_0y alors :
- E_x / E_0x = cos(theta) ;
- E_y / E_0y = sin(theta).

En combinant ces deux relations, on obtient donc l'équation d'un cercle :

![image](equation_cercle.png)

![image](polarisation_circulaire.png)

### Polarisation elliptique ###

C'est le résultat de la combinaison de la polarisation circulaire et de la polarisation linéaire.

![image](polarisation_elliptique.png)

### Lumière non polarisée ###

Il s'agit de la lumière naturelle.

LCP fait référence au champ électrique de polarisation ciculaire tournant dans le sens inverse des aiguilles d'une montre depuis le point de vue de l'observateur. 

Pour les diférents types de polarisation, on retrouve les relations suivantes :
- Polarisation linéaire : Q ≠ 0, U ≠ 0, V = 0 ;
- Polarisation circulaire : Q = 0, U = 0, V ≠ 0 ;
- Lumière entièrement polarisée : I^2 = Q^2 + U^2 + V^2 ;
- Lumière partiellement polarisée : I^2 > Q^2 + U^2 + V^2 ;
- Lumière non polarisée : Q = U = V = 0.

L'onde réflechie par un objet possède un cractère aléatoire. Pour la traiter, on calcule les covariances des composantes du champ électrique, ce qui nous permet de caractériser la polarisation de l'onde à l'aide du vecteur de Stokes. Les quantités mesurables sont les intensités des composantes du champ électrique. 

## Matrice de Mueller ##

La matrice de Mueller est un opérateur de taille 4x4 qui traduit l'action d'un système sur une onde incidente de vecteur de Stokes S_i.

L'opération est donc de la forme :

## Images multidimensionnelles de polarisation ##

Pour l'imagerie de Stokes, on a la mesure des vecteurs de Stokes pour chaque pixel d'une image ce qui donne une image de dimension 4.

Pour l'imagerie de Mueller, on a la mesure des matrices de Mueller pour chaque pixel d'une image ce qui donne une image de dimension 16.

### Acquisition des images de Stokes ###

On a un état incident et 4 états d'analyse ce qui donne 4 images. 

Le vecteur de Stokes peut également s'écrire de la façon suivante :

![image](vecteur_stokes_proprietes.png)

On retrouve :
- ρ = sqrt(S_1^2 + S_2^2 + S_3^2)/S_0, le degré de polarisation ;
- λ = 0.5 x tan^(-1){S_2 / S_1}, l'angle de polarisation ;
- φ = arg(S_1 + iS_2), l'orientation.

### Acquisition des images de Mueller ###

On a 4 états incidents et 4 états d'analyse pour chacun de ces états ce qui donne 16 images.


## Applications ##

L'imagerie polarimétrique est utilisée aux fins suivantes :
- Diagnostiques de maladies ;
- Contrôle qualité ;
- Enlever le brouillard d'une image ;
- Enlever la composante spéculaire d'une image (reflet) ;
- Reconstruction 3D d'objets transparents, fortement réfléchissants.
