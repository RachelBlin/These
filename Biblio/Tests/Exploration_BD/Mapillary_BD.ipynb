# Exploration de la BD Mapillary #

La BD Mapillary est une BD qui contient plus de 130 millions d'images de scènes routières prises dans tous les endroits du monde. Parmi ces images, on retrouve 25 000 images hautes résolutions segmentées à l'échelle du pixel en fonction des objets contenus dans l'image, soit 100 catégories. 

Les images de la BD ont été prises dans une grande variabilité de conditions météorologiques (soleil, pluie, neige, brouillard et brume) et à différents moments de la journée (aube, journée, crépuscule et nuit).

Dans ce notebook, je vais travailler sur un sous-échantillon de données de la BD.

Lien vers le site officiel : https://research.mapillary.com/

```python
import os
import imageio

# Structure des dossiers de la BD 

path = "/home/rblin/Documents/BD/Mapillary"

dossiers = os.listdir(path)
print("Contenu de la BD Mapillary : ", dossiers)

path_SA = path + "/" + dossiers[0]
path_O = path + "/" + dossiers[1]
path_Af = path + "/" + dossiers[2]
path_NA = path + "/" + dossiers[3]
path_As = path + "/" + dossiers[4]
path_E = path + "/" + dossiers[5]

# J'ai restructuré les dossiers au préalable pour séparer les images
# de leur ground truth

im_SA = os.listdir(path_SA + "/data")
gt_SA = os.listdir(path_SA + "/gt")
im_O = os.listdir(path_O + "/data")
gt_O = os.listdir(path_O + "/gt")
im_Af = os.listdir(path_Af + "/data")
gt_Af = os.listdir(path_Af + "/gt")
im_NA = os.listdir(path_NA + "/data")
gt_NA = os.listdir(path_NA + "/gt")
im_As = os.listdir(path_As + "/data")
gt_As = os.listdir(path_As + "/gt")
im_E = os.listdir(path_E + "/data")
gt_E = os.listdir(path_E + "/gt")

# Il y a deux exemples d'images pour chacun des pays

im_exemple = imageio.imread(path_SA + "/data/" + im_SA[0])
print("Taille des images : ", im_exemple.shape)
```

```python
import pylab as pl
import random

# Affichage des images de la BD 

pl.figure(1)
pl.figure(figsize=(15,30))
pl.clf()
for i in range(2) :
    pl.subplot(6,2,i+1)
    im_temp_path = path_SA + "/data/" + im_SA[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.subplot(6,2,i+3)
    im_temp_path = path_O + "/data/" + im_O[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.subplot(6,2,i+5)
    im_temp_path = path_Af + "/data/" + im_Af[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.subplot(6,2,i+7)
    im_temp_path = path_NA + "/data/" + im_NA[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.subplot(6,2,i+9)
    im_temp_path = path_As + "/data/" + im_As[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.subplot(6,2,i+11)
    im_temp_path = path_E + "/data/" + im_E[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
```

```python
# Affichage des images de la gt

pl.figure(2)
pl.figure(figsize=(15,30))
pl.clf()
for i in range(2) :
    pl.subplot(6,2,i+1)
    im_temp_path = path_SA + "/gt/" + gt_SA[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.subplot(6,2,i+3)
    im_temp_path = path_O + "/gt/" + gt_O[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.subplot(6,2,i+5)
    im_temp_path = path_Af + "/gt/" + gt_Af[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.subplot(6,2,i+7)
    im_temp_path = path_NA + "/gt/" + gt_NA[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.subplot(6,2,i+9)
    im_temp_path = path_As + "/gt/" + gt_As[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.subplot(6,2,i+11)
    im_temp_path = path_E + "/gt/" + gt_E[i]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
```
