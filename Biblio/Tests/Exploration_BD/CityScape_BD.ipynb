# Exploration de la BD City Scape #

CityScape est une BD qui permet d'illustrer la complexité des scènes urbaines. La segmentation est faite à l'échelle du pixel. Les images ont été tirées de sequencés vidéo prises dans les rues de 50 villes différentes. 5000 images ont une annotation de haute qualité à l'échelle du pixel et 20 000 images ont des annotations plus grossières. 

Lien vers le site officiel : https://www.cityscapes-dataset.com/

Lien vers l'article : https://arxiv.org/pdf/1604.01685.pdf

```python
# Exploration de la structure des dossiers

import os
import random
import imageio
import pylab as pl

# Structure des dossiers

path = "/home/rblin/Documents/BD/CityScape/leftImg8bit"

dossiers = os.listdir(path)
print("Nombre de sous-dossiers : ", len(dossiers), " Nom des sous-dossiers : ", print(dossiers))

path_test = path + "/" + dossiers[0]
path_train = path + "/" + dossiers[1]
path_val = path + "/" + dossiers[2]

# Nombre de sous-dossiers dans train 

sds = os.listdir(path_train)
print("Nombre de sous-dossiers : ", len(sds), " Nom des sous-dossiers : ", print(sds))

# Les dossiers sont donc organisés par ville.

# Nombre d'images dans la base de train
nb_im_train = 0
for i in sds :
    images = os.listdir(path_train + "/" + i)
    nb_im_train = nb_im_train + len(images)    

print("Nombre d'images de train total : ", nb_im_train) 

# Taille des images dans la BD 

path_imgs_example = path_train + "/" + sds[0]
images_example = os.listdir(path_imgs_example)
path_im_example = path_imgs_example + "/" + images_example[0]
im_example = imageio.imread(path_im_example)
print("Taille de l'image : ", im_example.shape)

# Affichage des paysages diversifiés de la BD train 

pl.figure(1)
pl.figure(figsize=(15, 20))
pl.clf()
for i in range(18) :
    pl.subplot(6,3,i+1)
    path_file_temp = path_train + "/" + sds[i]
    imgs_temp = os.listdir(path_file_temp)
    im_temp_path =  path_file_temp + "/" + imgs_temp[0]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.title(sds[i])
```

## Conclusion sur la BD ##

CityScape est une BD qui fournit de nombreuses informations sur les différents objets composant une image. Cependant, cette BD ne fournit pas d'images prises de nuit ou dans des conditions météorologiques dégradées, les images sont prises de jour dans des villes, et dans de bonnes conditions météorologiques.
