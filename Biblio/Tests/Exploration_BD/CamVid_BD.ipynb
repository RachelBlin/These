# Exploration de la BD CamVid #

La BD CamVid est la première BD qui regroupe des vidéos de scènes routières segmentées à l'échelle du pixel en 32 classes. Les images ont été labélisées à la main et il y en a 701 actuellement. 

Dans ce notebook, je vais travailler sur un sous-ensemble de 101 images de cette BD.

Lien vers le site officiel : http://mi.eng.cam.ac.uk/research/projects/VideoRec/CamSeq01/

```python
import os
import imageio

# Structure des dossiers de la BD

path = "/home/rblin/Documents/BD/CamVid"

dossiers = os.listdir(path)

# Je précise que j'ai resctruturé le dossier au préalable afin de 
# séparer la ground truth des images de celles-ci
print(dossiers)

path_data = path + "/" + dossiers[1]
path_gt = path + "/" + dossiers[3]

images = sorted(os.listdir(path_data))
gt = sorted(os.listdir(path_gt))

print("Nombre d'images : ", len(images))

im_test = imageio.imread(path_data + "/" + images[0])

print("Taille des images : ", im_test.shape)
```

```python
import pylab as pl
import random

# Affichage d'images au hasard de la BD 

indexes = random.sample(range(0,101), 9)

pl.figure(1)
pl.figure(figsize=(15,15))
pl.clf()
for i in range(9):
    pl.subplot(3,3, i+1)
    im_temp_path = path_data + "/" + images[indexes[i]]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.title(images[indexes[i]])
```

```python
# Affichage de la vérité terrain de ces images 

pl.figure(2)
pl.figure(figsize=(15,15))
pl.clf()
for i in range(9):
    pl.subplot(3,3, i+1)
    im_temp_path = path_gt + "/" + gt[indexes[i]]
    im_temp = imageio.imread(im_temp_path)
    pl.imshow(im_temp)
    pl.title(gt[indexes[i]])
```

## Conclusion sur la BD ##

Bien que la segmentation de cette BD soit très intéressante car elle indique de façon très précise les composants de l'image, elle est trop petite pour pouvoir effectuer un apprentissage robuste. De plus, les conditions météorologiques ne sont pas dégradées et il y a une faible variabilité en ce qui concerne les endroits où ont été prises les images.
